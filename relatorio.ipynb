# Análise Exploratória

Através da análise exploratória, vamos entender e **estruturar** os dados da forma adequada para a análise dos mesmos. 

Antes disso, é importante estabelecer algumas perguntas que podem ser respondidas com o desenrolar dessa análise:

    1. Existem relações entre a distorção e o tipo de zona  da escola?
    2. Existem relações entre a distorção e tipo de ensino (fundamental ou médio)?
    3. Existem relações entre a distorção e tipo de dependência administrativa (municipal, pública, estadual, federal ou privada)?

 **A análise irá se concentrar apenas nos dados de escolas do estado de Pernambuco**

### Importando o conjunto de dados

```python
import pandas as pd

df = pd.read_excel("datasets\TDI_ESCOLAS_2024.xlsx", skiprows=8)
df
```

Precisamos checar como o dataframe está inicialmente estruturado:

```python
df.info()
```

Agora vamos selecionar as colunas que importam para a nossa análise e apenas os registros ligados ao estado de Pernambuco:

```python
df_pe = df[df['SG_UF'] == 'PE'] 
```

```python
df_pe = df_pe.drop(columns=['FUN_AI_CAT_0','FUN_AF_CAT_0','FUN_01_CAT_0','FUN_02_CAT_0','FUN_03_CAT_0','FUN_04_CAT_0','FUN_05_CAT_0','FUN_06_CAT_0','FUN_07_CAT_0','FUN_08_CAT_0','FUN_09_CAT_0','MED_01_CAT_0','MED_02_CAT_0','MED_03_CAT_0','MED_04_CAT_0'])
df_pe
```


Dicionário dos dados atualizado:

* NO_REGIAO = Região aonde se encontra a escola analisada
* SG_UF = Estado da escola analisada
* CO_MUNICIPIO = Código do município da escola
* NO_MUNICIPIOI = Nome do município da escola
* CO_ENTIDADE = Nome da escola analisada
* NO_ENTIDADE = Código da escola analisada
* NO_CATEGORIA = Mostra a zona da escola analisada (Rural ou Urbana)
* NO_DEPENDENCIA = Mostra se a escola está administrada em âmbito municipal, estadual, federal ou é privada.
* FUN_CAT_0 = Taxa total para o Ensino Fundamental
* MED_CAT_0 = Taxa total para o Ensino Médio


Vamos considerar que os valores ausentes indicam que não existem escolas com aquele nível de ensino na região.



```python
df_pe = df_pe.replace('--', 0)
```

```python
df_pe
```

# Análise dos dados

Agora que os dados foram tratados e estruturados de forma mais adequada, vamos começar a analisá-los para tentar extrair informações para as perguntas que fizemos.

Vamos começar tentando buscar relações entre o tipo de zona em que a escola se encontra e a taxa de distorção:

```python
df_pe.groupby('NO_CATEGORIA')[['FUN_CAT_0', 'MED_CAT_0']].median().round(2)
```

Usamos a mediana, um tipo de medida de tendência central, para identificar qual tipo de zona está mais associada com a distorção, porém outros pontos devem ser levados em consideração:

* Existem valores faltantes, tanto para a taxa total do ensino fundamental quanto do ensino médio.
* Escolas em zonas rurais tendem a possuir menos estudantes do que as escolas de zonas urbanas.

Soluções:

* Vamos desconsiderar registros com valores nulos.
* Vamos deixar a parte a relação da zona da escola com as taxas de distorção.

```python
df_pe_semNulos = df_pe[(df_pe['FUN_CAT_0'] != 0) & (df_pe['MED_CAT_0'] != 0)]
```

```python
df_pe_semNulos
```

Vamos agora tentar relacionar o tipo de dependência com as taxas de distorção:

```python
df_pe_semNulos.groupby('NO_DEPENDENCIA')[['FUN_CAT_0', 'MED_CAT_0']].median().round(2)
```

A análise anterior indica algo que já era esperado: A condição financeira dos alunos é um fator que ajuda a diminuir a distorção idade-série.

Também ficou claro que as taxas de distorção são maiores durante o ensino médio do que no ensino fundamental. Isso pode ser ocasionado pelo fator da complexidade dos conteúdos vistos durante o Ensino Médio.

# Visualização dos dados

```python
import matplotlib.pyplot as plt

visual = df_pe_semNulos.groupby("NO_DEPENDENCIA")[["FUN_CAT_0", "MED_CAT_0"]].mean().plot(kind="barh", figsize=(10, 5))
visual.legend(["Taxa Total para o Ensino Fundamental", "Taxa Total para o Ensino Médio"])
plt.title("Mediana das taxas de distorção idade-série por Dependência Administrativa")
plt.ylabel("Tipo de administração")
plt.tight_layout()
plt.show()
```
